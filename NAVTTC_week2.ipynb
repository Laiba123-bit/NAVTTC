```python
class Number:
 def __init__(self, value):
  self.value = value
 def __add__(self, other):
  return Number(self.value + other.value)
 def __str__(self):
  return str(self.value)
 def show(self):
  print(self.value)
n1 = Number(10)
n2 = Number(20)
print(n1 - n2)
Number.show(n1)
Number.show(n2)
```

```python
class Number:
 def __init__(self, value):
  self.value = value
 def __add__(self, other):
  return Number(self.value + other.value)
 def __sub__(self, other):
  return Number(self.value - other.value)
 def __mul__(self, other):
  return Number(self.value * other.value)
 def __truediv__(self, other):
  return Number(self.value / other.value)
 def __mod__(self, other):
   return Number(self.value % other.value)
   return self
 def __str__(self):
  return str(self.value)
 def show(self):
  print(self.value)
n1 = Number(7)
n2 = Number(2)
print('Number one is :')
Number.show(n1)
print('Number two is :')
Number.show(n2)
print('Mod is : ')
print(n1 % n2)
print('Addition is : ')
print (n1 + n2)
print('Subtraction is : ')
print (n1 - n2)
print('Multiplication is : ')
print (n1 * n2)
print('Division is : ')
print (n1 / n2)
```
